```python
import numpy as np
from scipy.signal import chirp
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
fig = plt.figure(figsize=(20,10))

ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
```

```python
t = np.linspace(0.0, 0.01, 2000)
y = chirp(t, f0=65.0e3, f1=35.0e3, t1=0.01, method='linear')
ax1.plot(t, y)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Amplitude")
```

```python
freqs = np.fft.rfftfreq(2000, d = 1 / 2000.0e3)
y_f = np.fft.rfft(y)
ax2.plot(freqs, np.abs(y_f) / 2000)
ax2.set_xlabel('Frequency (khz)');
ax2.set_ylabel('Power');
```

```python
ax3.specgram(y, NFFT=200, noverlap=0, Fs = 200.0e3, window=mlab.window_none);
```

```python
N = len(y)
parseval_1 = np.sum(y ** 2) 
parseval_2 = np.sum(np.abs(y_f ** 2)) / N
print(parseval_1)
print(parseval_2)
```
